Buatlah sebuah game sederhana menggunakan bahasa python beserta algoritma dan flowchartnya!

```python
import random

#Kode untuk memasukkan daftar negara
negara = ['jepang', 'laos']

print("GAME TEBAK NEGARA")
print("Silakan tebak negara yang saya pilih.")

# Memilih sebuah negara secara acak
negara_rahasia = random.choice(negara)

# Loop untuk meminta input dari pemain
while True:
    tebakan = input("Masukkan tebakan Anda: ")

    # Memeriksa apakah tebakan pemain benar atau salah
    if tebakan.lower() == negara_rahasia:
        print("✅Anda benar! Negara yang saya pilih adalah", negara_rahasia)
        break
    else:
        print("❌Tebakan Anda salah. Coba lagi.")
```

---
> **ALGORITMA**
---
1.   Sistem akan memilih acak negara sesuai yang ada di data.
2.   Kemudian Pemain diharuskan menebak Negara yang dipilih oleh sistem, dengan memasukkan tebakkan kedalam kotak input.
3.   Setelah tebakkan dimasukkan maka akan menghasilkan 2 pernyataan yaitu :

*   Jika tebakkan benar maka akan keluar pernyataan "✅Anda benar! Negara yang saya pilih adalah (nama negara yang benar)".
*   Jika tebakkan salah maka akan keluar pernyataan "❌Tebakan Anda salah. Coba lagi.".
