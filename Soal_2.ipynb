Sebuah Toko sepatu menawarkan diskon 10% untuk pembelian sepatu dengan harga lebih dari Rp. 500.000 dan diskon 5% untuk pembelian sepatu lebih dari Rp. 100.000. Buatlah program python untuk menghitung harga akhir pembelian sepatu menggunakan kondisi IF, buatkan juga algoritma dan flowchart nya!

```python
#Tempat memasukkan total belanja
total_belanja = int(input('Total Belanja: Rp.'))

#Kode untuk mendapatkan POTONGAN DISKON 5%
if (total_belanja > 100000) and (total_belanja < 500000):
  harga_akhir = total_belanja - (0.05*total_belanja)
  print('Anda mendapat potongan diskon sebesar 5%')
#Kode untuk mendapatkan POTONGAN DISKON 10%
elif (total_belanja > 500000):
  harga_akhir = total_belanja - (0.1*total_belanja)
  print('Anda mendapat potongan diskon sebesar 10%')
#Kode untuk pernyataan lain
else:
  harga_akhir = total_belanja

#Kode untuk memunculkan hasil akhir
print('Total bayar: Rp.',round(harga_akhir,2))
```

---
> **ALGORITMA**
---
1.   Input Total Belanja Sepatu
2.   Setelah di Input, Maka akan menghasilkan 3 hasil/pernyataan yaitu :


*   Jika Total Belanja Rp. 100.000 ke bawah maka tidak akan mendapatkan POTONGAN DISKON, dan membayar sesuai Total Belanja.
*   Jika Total Belanja antara Rp. 101.000 s/d Rp. 500.000 maka akan mendapatkan POTONGAN DISKON sebesar 5%, dan Membayar sesuai dengan Total Bayar Akhir (Total Belanja - Diskon 5%).
*   Jika Total Belanja diatas Rp. 501.000 maka akan mendapatkan POTONGAN DISKON sebesar 10%, dan Membayar sesuai dengan Total Bayar Akhir (Total Belanja - Diskon 10%).
