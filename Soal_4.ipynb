```python
# Program mencetak bilangan ganjil dari 1 hingga N

# Input N dari pengguna

N = int(input("Masukkan nilai N: "))

# Mengecek apakah N positif
if N > 0:
    print("Bilangan ganjil dari 1 hingga", N, "adalah:")
    # Loop untuk mencetak bilangan ganjil
    for i in range(1, N + 1, 2):
        print(i)
```

---
> **ALGORITMA**
---
1.  Meminta pengguna untuk memasukkan nilai N.
2.  Sistem akan mencetak setiap bilangan ganjil yang ada diantara 0 sampai dengan BILANGAN yang di input pengguna."
