**CONDITIONAL STATEMENT**
---
Conditional statement digunakan untuk memeriksa suatu kondisi yang ditentukan sebelumnya apakah nilai nya True atau False.


```python
#Kode Untuk Pernyataan 1 Kondisi
angka = 300
if angka ==10:
  print("angka adalah 10")
```

```python
#Kode Untuk Pernyataan 2 Kondisi
angka = 200
if angka ==10:
  print("angka adalah 10")
else angka < 10:
  print("angka adalah kurang dari 10")
```

```python
#Kode Untuk Pernyataan Lebih dari 2 Kondisi
angka = 200
if angka ==10:
  print("angka adalah 10")
elif angka < 10:
  print("angka adalah kurang dari 10")
else :
  print("angka adalah lebih dari 10")
```

**FOR LOOPS**
---
Metode lain yang berguna dalam bahasa pemrograman Python adalah iterator. Hal ini sangat membantu jika Anda ingin melakukan sesuatu beberapa kali.
Kita dapat menggunakan suatu cara yang disebut loop atau iterator. Ada dua cara untuk menjalankan perintah tersebut, yaitu dengan for atau while loop.


```python
#Kode Menggunakan For Loop dengan penambahan Metode "range"
for i in range(3):
  print("Hai")
else:
  print("Selesai")
```

```python
#Kode Menggunakan For Loop dapat melakukan "nesting for loop"
for i in range(3):
     for j in range(2):
        print("loop di dalam")
     print("loop di luar")
```

```python
#Kode Menggunakan For Loop dengan statement "pass"
for i in range(3):
   pass
```

**WHILE LOOPS**
---
Dalam bahasa pemrograman Python, terdapat metode loop atau iterator yang lain, yaitu while loop. Kita pun dapat memberikan hasil yang sama seperti yang telah kita lakukan pada for loop dengan menggunakan while loop, seperti contoh di bawah ini:


```python
#Kode Menggunakan While Loop
i = 0
while i < 3:
  print("Angka", i)
  i += 1
```

```python
#Kode Menggunakan While Loop dengan penambahan statement "Break"
i = 0
while i < 3:
    if i == 3:
           break
    print("Angka", i)
    i += 1
```

```python
#Kode Menggunakan While Loop dengan penambahan statement "Else"
i = 0
while i < 3:
    print("Angka", i)
    i += 1
else:
    print("Angka lebih besar dari 2")
```

```python
#Kode Menggunakan While Loop dengan penambahan statement "Continue"
i = 0
while i < 5:
  i += 1
  if i == 2:
       continue
  print("Angka", i)
```
